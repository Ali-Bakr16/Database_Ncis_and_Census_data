```python
# Import the packages 
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load the first File
df1 = pd.read_csv('US_Census_Data.csv')
df1.head(10)
```

```python
df1.shape
```

This data contain 85 rows and 52 columns

```python
df1.info()
```

All columns at this data have object data type

```python
# Drop the Fact Note column Because it has more missing values
df1.drop(columns=['Fact Note'], inplace= True)
```

```python
df1.head()
```

```python
# Show number of missing values at every column
df1.isna().sum()
```

Every column has 20 missing value exepect the first column has only 5

```python
# Drop all missing values from data
df1.dropna(inplace= True)
```

```python
# Shape of data after droping missing values
df1.shape
```

After droping missing values the data has 65 rows and 51 columns

```python
#Remove comma from number to convert it to float
for i in df1.iloc[:,1:].columns:
    df1[i]=df1[i].str.replace(',','')

# Remove % sign from number to convert it to float 
for i in df1.iloc[:,1:].columns:
    df1[i]=df1[i].str.strip('%')   
    
for i in df1.iloc[:,1:].columns:
    df1[i]=df1[i].str.strip('$')
    
for i in df1.iloc[:,1:].columns:
    df1[i]=df1[i].str.strip('""')            

# Remove some letter from columns to convert to float
df1.iloc[42,:] = df1.iloc[42,:].replace('D','0')
df1.iloc[16,:] = df1.iloc[16,:].replace('Z','0')
df1.iloc[53,:] = df1.iloc[53,:].replace('FN','0')

# Convert the columns to float
for i in df1.iloc[:,1:].columns:
    df1[i] = df1[i].astype(float)
```

```python
df1.describe().round(1)
```

```python
# Load the second data
df2 = pd.read_excel('gun_data.xlsx')
df2.head()
```

```python
df2.info()
```

```python
# Show number of missing values at evey column
df2.isna().sum()
```

```python
# Drop cloumns that have missing values more than half 
df2.drop(columns=['permit_recheck','returned_handgun','returned_long_gun','returned_other','rentals_handgun','rentals_long_gun','private_sale_handgun','private_sale_long_gun','private_sale_other','return_to_seller_handgun','return_to_seller_long_gun','return_to_seller_other','prepawn_other','redemption_other','other'], inplace= True)
```

```python
df2.isna().sum()
```

```python
# Convert the month column to datetime data type
df2['month'] = pd.to_datetime(df2['month'])

# Create new column year that contain year value
df2['year'] = df2['month'].dt.year

# Create new column month that contain month value
df2['month'] = df2['month'].dt.month
```

```python
# Fill missing value by Zero
df2.fillna(0,inplace=True)
```

```python
df2.isna().sum()
```

```python
# Create new dataframe that contain all data about Alaska
df_Alaska = df2[df2['state'] == 'Alaska']
df_Alaska
```

```python
plt.bar(x=df_Alaska['month'], height=df_Alaska['totals'])
plt.xticks(np.arange(1,13))
plt.title('More Month Has total bought guns at Alasks')
plt.xlabel('Months')
plt.ylabel('Totals');
```

December is the month that has the largest sales

```python
plt.figure(figsize=(18,12))
plt.bar(x=df_Alaska['year'], height=df_Alaska['totals'])
plt.xticks(np.arange(1988,2018),rotation = 90)
plt.title('More year Has total bought guns at Alaska')
plt.ylabel('Totals')
plt.xlabel('Years');
```

2012 is the year that has the largest sales
